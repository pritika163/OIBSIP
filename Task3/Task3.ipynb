```python
df1=df
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
```

```python
cols=['fueltype','aspiration','doornumber','carbody','drivewheel', 'enginelocation','enginetype','cylindernumber','fuelsystem']
df1[cols]=df1[cols].apply(LabelEncoder().fit_transform)
df1.head()
```

```python
X=df1.drop(['CarName','price','car_ID'],axis=1)  #the independent variables
Y=df1['price']  #the dependent variable
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=10) #splitting the data into train and test set
```

```python
X_train.sort_index()  #displaying independent train set
```

```python
X_test.sort_index()  #displaying independent test set
```

```python
Y_train.sort_index()  #displaying dependent train set
```

```python
Y_test.sort_index()  #displaying dependent test set
```

```python
r=linear_model.LinearRegression()
r.fit(X_train,Y_train)
```

```python
re=r.predict(X_test)  #predicting the price
```

```python
re
```

```python
print(DataFrame({"Predicted price":re}))
```

```python
r.score(X_test,Y_test)  #gives the accuracy score of model
```
